# TASK 1


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# Fill missing values
df['age']=df['age'].fillna(df['age'].median())
df['city']=df['city'].fillna(df['city'].mode()[0])

# Remove duplicates
df= df.drop_duplicates()

print("Cleaned Data:\n",  df)
```

# TASK 2


```python
from sklearn.preprocessing import LabelEncoder

# One-Hot Encoding for city
df=pd.get_dummies(df,columns=['city'])

# Label Encoding for gender
le= LabelEncoder()
df['gender']=le.fit_transform(df['gender'])

print("\nEncoded Data:\n",df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

num_cols = ['age','annual_income']

# Min-Max Scaling
minmax=MinMaxScaler()
df_minmax=df.copy()
df_minmax[num_cols]=minmax.fit_transform(df[num_cols])

print("\nMin-Max Scaled:\n",df_minmax)

# Standardization
standard=StandardScaler()
df_standard=df.copy()
df_standard[num_cols]=standard.fit_transform(df[num_cols])

print("\nStandardized Data:\n",df_standard)
```

Min-Max Scaling is best when yu need values in a fixed range (like 0–1),especially for neural networks.
Standardization is better when data has outliers or when algorithms assume normal distribution (like SVM,Logistic Regression)
